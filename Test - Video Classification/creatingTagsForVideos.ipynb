```python
import pandas as pd

# open the .txt file which have names of all videos
f = open("classList.txt", "r")
temp = f.read()
videos = temp.split('\n')

# creating a dataframe having video names
all_vids = pd.DataFrame()
all_vids['video_name'] = videos
all_vids = all_vids[:-1]
all_vids.head()

# creating tags for all videos
video_tags = []
for i in range(all_vids.shape[0]):
    video_tags.append(all_vids['video_name'][i].split('/')[0])
    
all_vids['tag'] = video_tags
all_vids.shape[0]
all_vids
```
